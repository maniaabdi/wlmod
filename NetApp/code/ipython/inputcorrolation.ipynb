```python
import pandas as pd 
import numpy as np

df2 = pd.read_csv('code/input_access.csv')
print(df2.columns, len(df2))
```

```python
df = df2.head(20000)
print(df)
```

```python
files = [f.replace(' ', '').replace('\'', '') for f in df['fpath']]
u_files = [f.replace(' ', '').replace('\'', '') for f in df['fpath'].unique()]
n_files = df['fpath'].nunique()
```

```python
# corrolation matrix: with date 
icormat = np.full((n_files, n_files), -100)

icordict = {}
for i in range(0, n_files):
    fi = df.iloc[i]['fpath'].replace(' ', '').replace('\'', '')
    its = df.iloc[i]['timestamp']
    if fi not in icordict:
        icordict[fi] = {}
    print(i, n_files)
    for j in range(i, n_files):
        if i == j:
            continue
            
        fj = df.iloc[j]['fpath'].replace(' ', '').replace('\'', '')
        jts = df.iloc[j]['timestamp']
        
        if fj not in icordict[fi]: 
            icordict[fi][fj] = []
        icordict[fi][fj].append(jts - its)
        icormat[i][j] = round(sum(icordict[fi][fj]) / len(icordict[fi][fj]), 2)
        
        if fi == fj and i != j: 
            break
```
